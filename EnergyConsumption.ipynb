```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_usage_distribution(df):
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Usage_kWh'], bins=30, kde=True, color='skyblue')
    plt.title('Distribution of Usage (kWh)')
    plt.xlabel('Usage (kWh)')
    plt.ylabel('Frequency')
    plt.show()

def plot_usage_by_load_type(df):
    plt.figure(figsize=(10, 6))
    sns.violinplot(data=df, x='Load_Type', y='Usage_kWh', palette='Set2', inner='quartile')
    plt.title('Usage (kWh) by Load Type')
    plt.xlabel('Load Type')
    plt.ylabel('Usage (kWh)')
    plt.show()

def plot_usage_over_time(df):
    df_sample = df.sample(frac=0.1, random_state=42)  # Use a 10% sample for line plot
    plt.figure(figsize=(14, 6))
    sns.lineplot(data=df_sample, x='date', y='Usage_kWh', color='blue')
    plt.title('Usage (kWh) Over Time (Sampled)')
    plt.xlabel('Date')
    plt.ylabel('Usage (kWh)')
    plt.show()

plot_usage_over_time(df)
plot_usage_by_load_type(df)
plot_usage_distribution(df)
```

```python
from sklearn.preprocessing import StandardScaler

# Load the dataset and ensure date column is parsed correctly
df = pd.read_csv('Steel_industry_data.csv', parse_dates=['date'], index_col='date', dayfirst=True)

# Ensure the index is a DatetimeIndex
df.index = pd.to_datetime(df.index, dayfirst=True)


# Fill missing values if any
df.fillna(method='ffill', inplace=True)

# Convert categorical variables to numerical
df['WeekStatus'] = df['WeekStatus'].map({'Weekday': 0, 'Weekend': 1})
df['Day_of_week'] = df['Day_of_week'].map({
    'Monday': 0, 'Tuesday': 1, 'Wednesday': 2,
    'Thursday': 3, 'Friday': 4, 'Saturday': 5, 'Sunday': 6
})
df['Load_Type'] = df['Load_Type'].map({'Light_Load': 0, 'Medium_Load': 1, 'Maximum_Load': 2})

# Feature Engineering: Extract time-based features
df['hour'] = df.index.hour
df['day'] = df.index.day
df['month'] = df.index.month
df['year'] = df.index.year

# Drop the original 'NSM' column if it exists
if 'NSM' in df.columns:
    df.drop(columns=['NSM'], inplace=True)

# Normalize the data
# Define target variables
target_regression = df['Usage_kWh']  # Regression target
target_classification = df['Load_Type']  # Classification target

# Drop target columns from features
features = df.drop(columns=['Usage_kWh', 'Load_Type'])

# Apply StandardScaler only to features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Convert back to DataFrame
features_scaled = pd.DataFrame(features_scaled, columns=features.columns, index=df.index)

# Keep target columns as they are
df_scaled = features_scaled.copy()
df_scaled['Usage_kWh'] = target_regression  # Regression target remains unscaled
df_scaled['Load_Type'] = target_classification  # Classification target remains unscaled

print("Data after preprocessing:\n",df_scaled.head())
```

model='additive'
TimeSeries=Trend+Seasonality+Residual
1. Trend Component (Trend)
Represents the overall long-term increase or decrease in the data over time.
Examples: A steady increase in electricity usage due to population growth.
2. Seasonal Component (Seasonality)
Captures short-term, repeated patterns in the data at fixed intervals.
Examples:
Daily electricity consumption peaks during working hours and drops at night
3. Residual Component (Noise/Residual)
The remaining irregular fluctuations after removing the trend and seasonality.
Examples: Sudden power outages or unexpected weather conditions affecting energy consumption.

```python
# -------------------------------
# 4. Inefficiency Detection
# -------------------------------
def detect_inefficiencies(df):
    # Rule 1: Low Power Factor
    df['Low_Power_Factor'] = (df['Lagging_Current_Power_Factor'] < 0.9).astype(int)

    # Rule 2: High Reactive Power (Top 10%)
    reactive_threshold = df['Lagging_Current_Reactive.Power_kVarh'].quantile(0.9)
    df['High_Reactive_Power'] = (df['Lagging_Current_Reactive.Power_kVarh'] > reactive_threshold).astype(int)

    # Rule 3: Energy Spikes (Anomalies)
    from sklearn.ensemble import IsolationForest
    iso = IsolationForest(contamination=0.05, random_state=42)
    df['Energy_Spike'] = iso.fit_predict(df[['Usage_kWh']])
    df['Energy_Spike'] = np.where(df['Energy_Spike'] == -1, 1, 0)  # Convert to 0/1

    return df

# Apply to your dataset
df = detect_inefficiencies(df)

# Print inefficiency stats
print("\nInefficiency Statistics:")
print(f"Low Power Factor Cases: {df['Low_Power_Factor'].sum()}")
print(f"High Reactive Power Cases: {df['High_Reactive_Power'].sum()}")
print(f"Energy Spike Alerts: {df['Energy_Spike'].sum()}")
```

```python
# -------------------------------
# 5. Inefficiency Visualization
# -------------------------------
plt.figure(figsize=(14, 8))

# Plot 1: Low Power Factor
plt.subplot(2, 1, 1)
df['Lagging_Current_Power_Factor'].plot(alpha=0.7)
plt.axhline(0.9, color='red', linestyle='--', label='Power Factor Threshold (0.9)')
plt.title('Low Power Factor Detection')
plt.ylabel('Power Factor')
plt.legend()

# Plot 2: Energy Spikes
plt.subplot(2, 1, 2)
df['Usage_kWh'].plot(alpha=0.7, label='Energy Usage')
df[df['Energy_Spike'] == 1]['Usage_kWh'].plot(marker='X', linestyle='',
                                            color='red', label='Spike Detected')
plt.title('Energy Spike Detection')
plt.ylabel('Usage (kWh)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# 6. Actionable Insights
# -------------------------------
def generate_recommendations(df):
    # Insight 1: Peak hours with high reactive power
    reactive_peak_hours = df[df['High_Reactive_Power'] == 1].index.hour.value_counts().idxmax()

    # Insight 2: Days needing power factor correction (using value_counts)
    low_pf_days = df[df['Low_Power_Factor'] == 1].index.day_name().value_counts().idxmax()

    # Insight 3: High CO2 during energy spikes
    high_co2_spikes = df[df['Energy_Spike'] == 1]['CO2(tCO2)'].mean()

    print(f"""
    Actionable Recommendations:
    1. Schedule maintenance on {low_pf_days}s (most frequent low power factor days)
    2. Reduce operations during hour {reactive_peak_hours}:00 (peak reactive power)
    3. High CO2 levels ({high_co2_spikes:.2f} tCO2) detected during energy spikes
    """)
```
