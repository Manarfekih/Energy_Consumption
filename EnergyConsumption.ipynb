Visualizations

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_usage_distribution(df):
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Usage_kWh'], bins=30, kde=True, color='skyblue')
    plt.title('Distribution of Usage (kWh)')
    plt.xlabel('Usage (kWh)')
    plt.ylabel('Frequency')
    plt.show()

def plot_usage_by_load_type(df):
    plt.figure(figsize=(10, 6))
    sns.violinplot(data=df, x='Load_Type', y='Usage_kWh', palette='Set2', inner='quartile')
    plt.title('Usage (kWh) by Load Type')
    plt.xlabel('Load Type')
    plt.ylabel('Usage (kWh)')
    plt.show()

def plot_usage_over_time(df):
    df_sample = df.sample(frac=0.1, random_state=42)  # Use a 10% sample for line plot
    plt.figure(figsize=(14, 6))
    sns.lineplot(data=df_sample, x='date', y='Usage_kWh', color='blue')
    plt.title('Usage (kWh) Over Time (Sampled)')
    plt.xlabel('Date')
    plt.ylabel('Usage (kWh)')
    plt.show()

plot_usage_over_time(df)
plot_usage_by_load_type(df)
plot_usage_distribution(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_usage_by_time_components(df):
    # Convert Date column to datetime
    df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y %H:%M')

    # Extract time components
    df['Hour'] = df['date'].dt.hour
    df['Month'] = df['date'].dt.month
    df['Year'] = df['date'].dt.year

    plt.figure(figsize=(14, 6))

    # Plot 1: Usage by Hour of the Day
    plt.subplot(1, 3, 1)
    sns.lineplot(data=df, x='Hour', y='Usage_kWh', estimator='mean', ci=None, marker='o')
    plt.title('Usage (kWh) by Hour of Day')
    plt.xlabel('Hour')
    plt.ylabel('Usage (kWh)')

    # Plot 2: Usage by Month
    plt.subplot(1, 3, 2)
    sns.barplot(data=df, x='Month', y='Usage_kWh', estimator='mean', ci=None, palette='Blues')
    plt.title('Usage (kWh) by Month')
    plt.xlabel('Month')
    plt.ylabel('Usage (kWh)')


plot_usage_by_time_components(df)
```

we will replace the type of 'object column' by Dummy Variables

```python
from sklearn.preprocessing import StandardScaler

# Load the dataset and ensure date column is parsed correctly
df = pd.read_csv('Steel_industry_data.csv', parse_dates=['date'], index_col='date', dayfirst=True)

# Ensure the index is a DatetimeIndex
df.index = pd.to_datetime(df.index, dayfirst=True)


# Fill missing values if any
df.fillna(method='ffill', inplace=True)

# Convert categorical variables to numerical
df['WeekStatus'] = df['WeekStatus'].map({'Weekday': 0, 'Weekend': 1})
df['Day_of_week'] = df['Day_of_week'].map({
    'Monday': 0, 'Tuesday': 1, 'Wednesday': 2,
    'Thursday': 3, 'Friday': 4, 'Saturday': 5, 'Sunday': 6
})
df['Load_Type'] = df['Load_Type'].map({'Light_Load': 0, 'Medium_Load': 1, 'Maximum_Load': 2})

# Feature Engineering: Extract time-based features
df['hour'] = df.index.hour
df['day'] = df.index.day
df['month'] = df.index.month
df['year'] = df.index.year

# Drop the original 'NSM' column if it exists
if 'NSM' in df.columns:
    df.drop(columns=['NSM'], inplace=True)

# Normalize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Convert back to DataFrame
df_scaled = pd.DataFrame(df_scaled, columns=df.columns, index=df.index)

print("Data after preprocessing:\n", df_scaled.head())
```

model='additive'
TimeSeries=Trend+Seasonality+Residual
1. Trend Component (Trend)
Represents the overall long-term increase or decrease in the data over time.
Examples: A steady increase in electricity usage due to population growth.
2. Seasonal Component (Seasonality)
Captures short-term, repeated patterns in the data at fixed intervals.
Examples:
Daily electricity consumption peaks during working hours and drops at night
3. Residual Component (Noise/Residual)
The remaining irregular fluctuations after removing the trend and seasonality.
Examples: Sudden power outages or unexpected weather conditions affecting energy consumption.

# **Data distribution**

```python
from sklearn.model_selection import train_test_split

# Separate features and target variables
X = df.drop(columns=['Usage_kWh', 'Load_Type'])  # Features for regression and classification
y_reg = df['Usage_kWh']  # Target for regression
y_clf = df['Load_Type']  # Target for classification

# Split the data into training and testing sets
X_train, X_test, y_train_reg, y_test_reg, y_train_clf, y_test_clf = train_test_split( df_scaled , y_reg, y_clf, test_size=0.2, random_state=42, shuffle=False)
```

# ** Regression (Predict Energy Consumption)**

Using Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report
import numpy as np

# Train a Random Forest Regressor
reg_model = RandomForestRegressor(n_estimators=100, random_state=42)
reg_model.fit(X_train, y_train_reg)

# Make predictions
y_pred_reg = reg_model.predict(X_test)

# Evaluate the model
rmse = np.sqrt(mean_squared_error(y_test_reg, y_pred_reg))
print(f'RMSE for Regression: {rmse}')

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(y_test_reg.index, y_test_reg, label='Actual')
plt.plot(y_test_reg.index, y_pred_reg, label='Predicted')
plt.legend()
plt.title('Energy Consumption Prediction')
plt.show()
```

# Classification (Identify Load Type or Inefficiencies)

Random Forest Classifier

```python
# Import the Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

# Initialize the model
clf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model on the training data
clf_model.fit(X_train, y_train_clf)

# Make predictions on the test data
y_pred_clf = clf_model.predict(X_test)

# Evaluate the model using Accuracy and Classification Report
accuracy = accuracy_score(y_test_clf, y_pred_clf)
print(f'Accuracy for Random Forest Classification: {accuracy}')
print("Classification Report:")
print(classification_report(y_test_clf, y_pred_clf))

# Plot feature importance
importances = clf_model.feature_importances_
feature_names = X_train.columns
plt.figure(figsize=(10, 6))
plt.barh(feature_names, importances)
plt.xlabel('Feature Importance')
plt.title('Feature Importance for Classification (Random Forest)')
plt.show()
```
